# 막대 그래프

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('data/sports.csv', index_col = 0)
df
```

```python
df.plot(kind='bar')
```

```python
# 가로 막대 그래프
df.plot(kind='barh')
```

```python
# 누적 막대 그래프
df.plot(kind='bar', stacked = True)
```

```python
# 여성을 대상으로한 조사 결과
df['Female'].plot(kind='bar')
```
