## groupby

```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('data/laptops.csv')
df
```

```python
brand_nation = {
'Dell': 'U.S.',
'Apple': 'U.S.',
'ACer': 'Taiwan',
'HP': 'U.S.',
'Lenovo': 'China',
'Alienware': 'U.S.',
'Microsoft': 'U.S.',
'Asus': 'Taiwan'
}
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
df
```

```python
nation_groups = df.groupby('brand_nation')   
```
