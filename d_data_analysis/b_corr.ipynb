## 상관 관계 분석 (Correlation Analysis)

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/young_survey.csv")
df
```

```python
# 음악 장르
music = df.iloc[:, :19]
music.head()
```

```python
# 히트맵
sns.heatmap(music.corr(numeric_only=True))
```

```python
# 나이에 대한 상관 관계 분석
df.corr(numeric_only=True)['Age'].sort_values(ascending=False)
```
