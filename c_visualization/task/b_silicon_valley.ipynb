실리콘 밸리에서 일하는 사람들의 정보가 있습니다. 
직업 종류, 인종, 성별 등이 포함되어 있는데요.
실리콘 밸리에서 일하는 남자 관리자 (Managers)에 대한 인종 분포를 막대 그래프로 다음과 같이 그려보세요. 

```python
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('../data/silicon_valley_summary.csv')
df
```

```python
# 필터링
filtered_df = df[(df['gender'] == 'Male') & (df['job_category'] == 'Managers') & (df['race_ethnicity'] != 'All')]
```

```python
# 막대 그래프 그리기
filtered_df.plot(kind='bar', x='race_ethnicity',  y='count')
```
