## KDE 활용 예시

```python
import pandas as pd
import seaborn as sns
```

```python
body_df = pd.read_csv('data/body.csv', index_col=0)
body_df
```

```python
# Height 컬럼에 대한 히스토그램
body_df.plot(kind='hist', y='Height')
```

```python
# 히스토그램 위에 kde 곡선
sns.distplot(body_df['Height'])
```

```python
# Height 컬럼에 대한 상자 그림
body_df.plot(kind='box', y='Height')
```

```python
# 바이올린 플롯: 분포 전체를 보여줌
sns.violinplot(y=body_df['Height'])
```

```python
# Height와 Weight에 대한 산점도
body_df.plot(kind='scatter', x='Height', y='Weight')
```

```python
# x축은 Height, y축은 Weight 3차원
sns.kdeplot(body_df['Height'], body_df['Weight'])
```
