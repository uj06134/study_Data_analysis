## 클러스터 분석(Cluster Analysis)

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/young_survey.csv")
df
```

```python
# 관심사에 대한 컬럼
interests = df.loc[:, 'History':'Pets']
interests.head()
```

```python
corr = interests.corr()
corr
```

```python
# 역사와 연관성이 높은 관심사
corr['History'].sort_values(ascending=False)
```

```python
# 클러스터맵
sns.clustermap(corr)
```
